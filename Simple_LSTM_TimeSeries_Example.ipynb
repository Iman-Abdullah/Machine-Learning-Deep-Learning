```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
# 1. Generate dummy time series data (sine wave + noise)
timesteps = 200
x = np.arange(timesteps)
y = np.sin(0.1 * x) + np.random.normal(0, 0.1, size=timesteps)
```

```python
# 2. Prepare data for LSTM (sliding windows)
X = []
Y = []
window_size = 10
for i in range(len(y) - window_size):
    X.append(y[i:i+window_size])
    Y.append(y[i+window_size])
X = np.array(X)
Y = np.array(Y)
```

```python
# Reshape X for LSTM [samples, timesteps, features]
X = X.reshape((X.shape[0], X.shape[1], 1))
```

```python
# 3. Build LSTM model
model = Sequential([
    LSTM(50, activation='relu', input_shape=(window_size, 1)),
    Dense(1)
])
```

```python
# 4. Compile and train
model.compile(optimizer='adam', loss='mse')
model.fit(X, Y, epochs=50, verbose=0)
```

```python
# 5. Predict
y_pred = model.predict(X)
```

```python
# 6. Visualize
plt.plot(Y, label="True")
plt.plot(y_pred, label="Predicted")
plt.title("Dummy Time Series Forecasting with LSTM")
plt.legend()
plt.show()
```
