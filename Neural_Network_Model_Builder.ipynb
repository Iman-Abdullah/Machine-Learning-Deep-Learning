# Neural Network Dummy Data

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
# Create dummy data
x = np.linspace(0, 10, 100).reshape(-1, 1)        # 100 values from 0 to 10
y = 3 * x + 2 + np.random.normal(0, 1, size=x.shape)  
```

```python
# Build the model using Functional API
input_layer = Input(shape=(1,))
hidden = Dense(10, activation='relu')(input_layer)
output = Dense(1)(hidden)
```

```python
model = Model(inputs=input_layer, outputs=output)
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mse')
```

```python
# Train the model
history = model.fit(x, y, epochs=100, verbose=0)
```

```python
# Predict
x_test = np.array([[5.0]])
y_pred = model.predict(x_test)
print(y_pred)
```

```python
# Visualize results
y_fit = model.predict(x)
```

```python
plt.scatter(x, y, label='True Data')
plt.plot(x, y_fit, color='red', label='Model Prediction')
plt.title("Neural Network Fit to Dummy Data")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
